```python
Text_words = re.findall(r'\w+', Text)
split_regex = re.compile(r'[.|!|?|…]')
sentences = filter(lambda t: t, [t.strip() for t in split_regex.split(Text)])
```

```python
count_sentence = 0
count_syllable = 0
countWordInSentence = np.zeros(0, dtype=np.int)
```

```python
for sentence in sentences:
    count_sentence += 1
    words = re.findall(r'\w+', sentence)
    countWordInSentence = np.append(len(words), countWordInSentence)
    for s in re.findall('[бвгджзйклмнпрстфхцчшщьъ]*?[аеёиоуыэюя]', sentence, re.I):
#        print(s)
        count_syllable += 1
#    print(i, '=== ', s)
```
