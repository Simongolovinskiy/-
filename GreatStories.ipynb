### Инфинитив слова и отбросить его приставку

```python
# stack
# инфинитив слова и отбросить его приставку
morph = MorphAnalyzer()

def cut_prefix(word, morph=MorphAnalyzer()):
    analyzers = [
        KnownPrefixAnalyzer(morph), 
        UnknownPrefixAnalyzer(morph)]
    seen = set()
    word_lower = word.lower()
    res = []
    for analyzer in analyzers:
        parsed = analyzer.parse(word, word_lower, seen)
        if not parsed:
            continue
        tmp = parsed[0]
        res.extend(tmp)
        if len(tmp) >= 5:
            if tmp[4] and tmp[4][0]:
                return morph.normal_forms(tmp[4][0][1])[0]
    return res
```

### Кол-во символов в тексте

```python
def get_CountSymbols(text: str) -> str:
# Функция вычисляет колличество символов в тексте
    Text_clr_paragraph = text.replace('\n',' ')
    symbols = len(Text_clr_paragraph)
    return symbols
# get_CountSymbols(Text) 
```

### Кол-во букв в тексте

```python
def get_CountLetters(text: str) -> str:
# Функция вычисляет колличество букв в тексте
    text = re.sub('[^a-zа-яё]', '', text, flags=re.IGNORECASE)
    letters = len(text)
#     word = word.strip('-')
    return letters
# get_CountLetters(Text) 
```

### Кол-во слогов в тексте

```python
def get_CountStyle(text: str) -> str:
# Функция вычисляет колличество слогов в тексте
    style_ru = 0
    vowels = 'аоуэыияюёе'
    for leter in vowels:
        style_ru  += text.count(leter)
    return style_ru
# get_CountStyle(Text)
```

### Кол-во параграфов в слов в тексте и выдает список

```python
def get_CountWords(text: str) -> str:
# Функция вычисляет колличество слов в тексте и выдает их список 
    symbols = ['—', '\n', ',', '!', '.', ':', ';', '?']
    filtr = text
    for smb in symbols:
        filtr = filtr.replace(smb,' ')
    filtr = filtr.replace('-','')
    filtr = re.sub('\s+', ' ', filtr)
    filtr = filtr.lower()
    count_words = len(filtr.split())
    words = filtr.split()
    return count_words, words
count_words, words = get_CountWords(Text)
# count_words
# dic_words = (list(set(words)))
# dic_words.sort()
# dic_words
# words
```

### Кол-во параграфов в тексте

```python
def get_Paragraph(text: str) -> str:
    count_paragraph = 0
    symbls = ['—', '-', '*']
    paragraph = ['']    
    for pargrh in re.split( '\n', Text): 
        if pargrh !='':
            if pargrh[0] not in symbls:
                paragraph.append(pargrh)
                count_paragraph += 1
            else:
                paragraph[count_paragraph] += pargrh
    if paragraph[0] == '': 
        paragraph.remove('')
    return paragraph, count_paragraph
paragraph, count_paragraph = get_Paragraph(Text)
# get_Paragraph(Text)
(list(paragraph))
```
